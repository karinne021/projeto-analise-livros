### Tratando os dados nulos e faltantes

```python
# Substitui o texto [] por 'Sem Info'
df = df.replace('[]','No Info')
```

### Tratando Datas

```python
# Alterando as colunas para o formato Data

df['publishDate'] = pd.to_datetime(df['publishDate'], errors='coerce')

df['firstPublishDate'] = pd.to_datetime(df['firstPublishDate'], errors='coerce')
```

### Tratando o valor inválido da coluna isbn

```python
# Substitui o valor 9999999999999 por 'Sem Info'
df['isbn'] = df['isbn'].replace('9999999999999','No Info')
```

### Análise de Avaliações

```python
# Media das avaliações
media_rating = df['rating'].mean()
print(f"Média das Avaliações: {media_rating:.2f}")
```

```python
# Criar o histograma
n, bins, patches = plt.hist(df['rating'], bins=5, color='mediumpurple', edgecolor='black') 

# Adicionar valor sobre as barras
for i in range(len(patches)):
    height = n[i]  
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, height + 0.5,  
             str(int(height)),  
             ha='center',       
             fontsize=10) 


plt.title("Distribuição das Avaliações")
plt.xlabel("Avaliações")
plt.ylabel("Frequência")


# Salvar gráfico no diretório
plt.savefig('output/grafico_analise_avaliacoes.png')

plt.show()
```

### 10 Livros mais bem avaliados

```python
# 10 livros mais bem avaliadas considerando total de avaliações e nota
rating_books = df.sort_values(by=['numRatings','rating'], ascending=[False, False]).reset_index(drop=True).head(10)
```

```python
rating_books
```

```python
# Plotar as principais avaliações

# Ordenar os dados
dados_ordenados = sorted(zip(rating_books['rating'], rating_books['title']), reverse=True)
y_ordenado, x_ordenado = zip(*dados_ordenados)


# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x_ordenado, y_ordenado, color='navy')

# Adicionar título e rótulos
plt.title("Melhores avaliações", fontsize=16)
plt.xlabel("Livros", fontsize=12)
plt.ylabel("Avaliação", fontsize=12)


# Ajustar os rótulos do eixo x
plt.xticks(rotation=80, fontsize=10)

# Adicionar valores no topo das barras
for i, value in enumerate(y_ordenado):
    plt.text(i, value + 0.05,  
             str(value),       
             ha='center',      
             fontsize=10)  
    
# Ajustar margens
plt.margins(y=0.2)


# Salvar gráfico no diretório
plt.savefig('output/grafico_melhores_avaliacoes.png')

# Mostrar o gráfico
plt.show()
```

### Correlação das colunas com Avaliação

```python
# Seleciona apenas as colunas numéricas
df_numerico = df.select_dtypes(include=['number'])

# Calcula a correlação entre as colunas numéricas
correlacoes = df_numerico.corr()

# Escolhe a coluna de interesse para analisar a correlação
coluna_interesse = 'rating'

# Seleciona as variáveis mais correlacionadas com a coluna de interesse
corr = correlacoes[coluna_interesse].abs().sort_values(ascending=False)

# Exclui a própria coluna de interesse (pois sua correlação com ela mesma é 1)
corr = corr.drop(coluna_interesse)


# Cria um subset do DataFrame com as 10 variáveis mais correlacionadas
variaveis_corr = corr.index.tolist()

# Cria a matriz de correlação apenas com as 10 variáveis mais correlacionadas
matriz_corr = correlacoes.loc[variaveis_corr, variaveis_corr]


print(f"Fatores mais correlacionados com a Avaliação dos Livros:\n\n {matriz_corr.index.tolist()}")
```

```python
# Plota a matriz de correlação das 10 principais variáveis
plt.figure(figsize=(10, 6))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Variáveis Mais Correlacionadas com a Avaliação dos Livros")

# Salvar gráfico no diretório
plt.savefig('output/grafico_correlacao_avaliacoes.png')

plt.show()
```

### Análise de Lançamento dos Livros

```python
# Realiza uma cópia do dataframe
df_lancamento = df.copy()
```

```python
# Extraindo o ano
df_lancamento['publishDate_year'] = pd.to_datetime(df_lancamento['publishDate']).dt.year
```

```python
# Selecionar somente até o ano atual
livros_ano = df_lancamento[df_lancamento['publishDate_year'] <= 2025]
```

```python
# Criar índice
livros_ano = livros_ano['publishDate_year'].value_counts().sort_index()
```

```python
# Plotar gráfico
plt.figure(figsize=(10,6))
livros_ano.plot(kind='line', color='navy')

plt.title('Publicações de Livros ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Publicações')
plt.xticks(rotation=45)


# Salvar gráfico no diretório
plt.savefig('output/grafico_publicacoes_anos.png')

plt.show()
```
